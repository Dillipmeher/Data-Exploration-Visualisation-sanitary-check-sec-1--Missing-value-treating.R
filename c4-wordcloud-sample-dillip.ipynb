## Preparing Data for Wordcloud

```python
#read the course descriptions
with open("/kaggle/input/text-handling-ipba-kn/1_TextRepresentation/data_science.txt", 'r') as fh:  
    filedata = fh.read()

#check contents
print("File data sample : ", filedata[:200])
```

```python
# Create stopword list:
from wordcloud import WordCloud, STOPWORDS
stopwords = set(STOPWORDS)

#Generate wordcloud data
wordcloud = WordCloud(stopwords=stopwords, max_words=25, \
                      background_color="white").generate(filedata)
```

```python
wordcloud
```

## Displaying the Wordcloud

```python
import matplotlib.pyplot as mpLib
mpLib.imshow(wordcloud)
mpLib.axis("off")
mpLib.show()
```

## Enhancing Wordcloud (adding custom words to stopwords)

```python
 # Check stopwords
stopwords
```

```python
#Add more words to ignore 
stopwords.update(["helps","study", "fields", "happened"])
```
